```python
# 定義：以 % 或 %% 為前綴的特殊命令，可提供增強的功能
%timeit x=[ i for i in range(1000)] #當行程式
```

```python
# 多行程式
%%timeit
total = 0
for i in range(1000):
    total += i
```

```python
# 圖形：使用 Matplotlib 或 Seaborn 進行可視化。
# 表格：使用 Pandas 建立和顯示表格。

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataframe
Data = {'Name' : ['Alice', 'Bob', 'Charlie', 'David'],
 'Age' : [24,27,22,32]}

df = pd.DataFrame(Data)

# Display Table
print(df)
```
