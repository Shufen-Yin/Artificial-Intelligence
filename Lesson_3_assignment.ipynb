```python
# 9. Web API Interaction (GitHub API)
def get_python_repos():
    try:
        url = "https://api.github.com/search/repositories"
        params = {"q": "language:python", "sort": "stars", "order": "desc"}
        response = requests.get(url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes
        data = response.json()
        return [repo['name'] for repo in data['items'][:5]]
    except requests.RequestException as e:
        print(f"Error fetching data from GitHub API: {e}")
        return []

# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())
```

```python
# Additional Features (Bonus)

# Feature 1: Remove book from library
def remove_book(book_name):
    if book_name in books:
        books.remove(book_name)
        print(f"Book '{book_name}' removed from library.")
    else:
        print(f"Book '{book_name}' not found in library.")

# Feature 2: Update book information in database
def update_book_quantity(book_id, new_quantity):
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()
        cursor.execute("UPDATE books SET quantity = ? WHERE id = ?", (new_quantity, book_id))
        conn.commit()
        conn.close()
        print(f"Book ID {book_id} quantity updated to {new_quantity}.")
    except sqlite3.Error as e:
        print(f"Error updating book: {e}")

# Feature 3: Simple command-line interface
def library_menu():
    while True:
        print("\n=== Library Management System ===")
        print("1. View all books")
        print("2. Check book availability")
        print("3. Add new book")
        print("4. Remove book")
        print("5. View books in database")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            print("\nAvailable books:")
            for i, book in enumerate(books, 1):
                print(f"{i}. {book}")

        elif choice == '2':
            book_name = input("Enter book name to check: ")
            available = check_availability(book_name)
            print(f"Book '{book_name}' is {'available' if available else 'not available'}.")

        elif choice == '3':
            new_book = input("Enter new book title: ")
            books.append(new_book)
            print(f"Book '{new_book}' added to library.")

        elif choice == '4':
            book_name = input("Enter book name to remove: ")
            remove_book(book_name)

        elif choice == '5':
            db_books = display_books_from_db()
            print("\nBooks in database:")
            for book in db_books:
                print(f"ID: {book[0]}, Title: {book[1]}, Quantity: {book[2]}")

        elif choice == '6':
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

# Uncomment the line below to test the menu interface
# library_menu()

print("\nLibrary Management System operations completed.")
```
