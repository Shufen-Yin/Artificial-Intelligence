```python
# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:
# Create tuple and set
fruits_tuple = ('apple', 'banana', 'orange')
numbers_set = {1, 2, 3, 4, 5}

print(f"Original tuple: {fruits_tuple}")
print(f"Original set: {numbers_set}")

# Try to add elements
try:
    # Tuples are immutable - this will raise an error if we try to modify in-place
    fruits_tuple += ('grape',)  # This actually creates a new tuple
    print(f"After += operation: {fruits_tuple}")
except Exception as e:
    print(f"Cannot modify tuple directly: {e}")

# Alternative way to demonstrate tuple immutability
try:
    # This will definitely fail - trying to modify tuple element directly
    fruits_tuple[0] = 'kiwi'
except TypeError as e:
    print(f"Cannot modify tuple elements: {e}")

# Create a new tuple with additional element
new_fruits_tuple = fruits_tuple + ('grape',)
print(f"New tuple after addition: {new_fruits_tuple}")

# Sets are mutable - we can add elements directly
numbers_set.add(6)
print(f"Modified set after adding 6: {numbers_set}")

# Add another element to set
numbers_set.add(7)
print(f"Modified set after adding 7: {numbers_set}")

print("\nExplanation:")
print("- Tuples are immutable: cannot be changed after creation")
print("- Sets are mutable: can add/remove elements after creation")
print("- When we use += with tuples, it creates a new tuple object")
print("- Sets use methods like add() to modify the original set")
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Iterator class for even numbers
class EvenNumberIterator:
    def __init__(self, count):
        self.count = count
        self.current = 0
        self.generated = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.generated >= self.count:
            raise StopIteration
        result = self.current
        self.current += 2
        self.generated += 1
        return result

# Generator function for even numbers
def even_number_generator(count):
    current = 0
    generated = 0
    while generated < count:
        yield current
        current += 2
        generated += 1

print("Using iterator:")
even_iterator = EvenNumberIterator(5)
for num in even_iterator:
    print(num, end=" ")

print("\nUsing generator:")
even_gen = even_number_generator(5)
for num in even_gen:
    print(num, end=" ")
print("\n")
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:

# Map - square all numbers
squared_numbers = list(map(lambda x: x ** 2, numbers))

# Reduce - product of all numbers
product = reduce(lambda x, y: x * y, numbers)

# Filter - even numbers only
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(f"Original numbers: {numbers}")
print(f"Squared numbers (map): {squared_numbers}")
print(f"Product of all numbers (reduce): {product}")
print(f"Even numbers (filter): {even_numbers}")
print()
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:

class Rectangle:
    def __init__(self, length, width):
        # Validate inputs
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive numbers")
        self.length = length
        self.width = width

    def area(self):
        """Calculate and return the area of the rectangle."""
        return self.length * self.width

    def perimeter(self):
        """Calculate and return the perimeter of the rectangle."""
        return 2 * (self.length + self.width)

    def __str__(self):
        return f"Rectangle(length={self.length}, width={self.width})"

# Create rectangle objects
rect1 = Rectangle(5, 3)
rect2 = Rectangle(7, 4)

print(f"Rectangle 1: {rect1}")
print(f"Area: {rect1.area()}")
print(f"Perimeter: {rect1.perimeter()}")

print(f"\nRectangle 2: {rect2}")
print(f"Area: {rect2.area()}")
print(f"Perimeter: {rect2.perimeter()}")
print()
```
